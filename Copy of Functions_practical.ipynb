```python
#Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_list(n):
    sum=0
    for i in n:
        if i%2==0:
            sum+=i
    return sum
sum_list([1,2,3,4,5,6])
```

```python
#Write a Python function that checks if a given number is prime or not from 1 to 200

import math
def is_prime(num):
    if num<=1:
        return False
    for i in range(2,int(math.sqrt(num))+1):
        if num%i==0:
            return False
    return True

for i in range(1,201):
    if is_prime(i):
        print(f"{i} is prime")
```

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fib(num):
    a=0
    b=1
    for i in range(num):
        yield a
        a,b=b,a+b
s=fib(1000)
for i in range(6):
    print(next(s), end=" ")
```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent

def sqr(num):
    for i in range(1, num+1):
        yield i**2
s=sqr(1000)
for i in range(1,8):
    print(f"square of {i}: {next(s)}",end="\n")
```

```python
#Implement a generator function that reads a file line by line and yields each line as a string.

def read_lines(s):
    for i in s.split("\n"):
        yield i
file="""line 1: Hello
line 2: This is Test
line 3: Generator Function
line 4: Thank you
"""
gen=read_lines(file)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#Write a python program that uses 'filter()' to remove all vowels from a given string.
s="hello world"
vowel=["a","e","i","o","u","A","E","I","O","U"]
result=list(filter(lambda x:x not in vowel,s))
print("original:",s)
print("without vowel:")
print(''.join(result)) #join helps to converts the filtered characters back into a string.
```

```python
"""
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.
"""


orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]
product =[]
for price in orders:
    product.append(price[2]*price[3])

order_total=list(map(lambda x,y:(x[0],y + (10 if y<=100 else 0)), orders, product))

print(order_total)
```
